## 2. Basic stats and understanding the data

### 2.1 Choices in data cleaning and preprocessing

Dataset 1 had some empty values within the column "samlet_opvarmet_areal". Those were removed from the database. 
Other than that, some of the cities mentioned in dataset 1, was not mentioned in dataset 2. Those were removed from the dataframe as well. 
<br><br>
The naming of some cities was also different in dataset 1 compared to dataset 2. For example, was København K in dataset 1, named as Indre by in dataset 2. The city names were changed accordingly. 
<br><br>
Dataset 2 did not have any empty cells, but the information given about city name, gender and civil status was given in number. The explanation  of what the number means was given on their website. For example, did number 1 for city name mean "Indre By". All those number for each column replaced with real names. Dataset 2 also included cities, which did not have a specific name. This data was removed. Lastly, when doing the visualization of age groups a category name was created for each row, explaining which group it belonged to.

## 4.Genre


### 4.1 Which tools are used from each of the 3 categories of Visual Narrative (Figure 7 in Segal and Heer) and why?

Magazine genre is selected, I wanted to deliver the answer like it was a report, but on the other hand I wanted them to interact and learn from the data as well. The visualizations are therefore interactive.

Following tools are used: 
* Consistent Visual Platform - because the layout of a magazine report should be consistent
* Zooming - to make it interactive and let the user learn from the data
* Feature distinction - to have different answers in different places in the report

### 4.2 Which tools are used from each of the 3 categories of Narrative Structure (Figure 7 in Segal and Heer) and why?

Following tools are used: 
* Random Access - the user can skip the answer of where the heat savings potential is in the city or target group by scrolling
* Hover Highlighting / Details - to make it interactive and let the user learn from the data
* Filtering / Selection / Search - to make it interactive and let the user learn from the data
* Captions / Headlines - A report must describe each section
* Introductory Text - to explain the goal of the report and what it is about
* Summary / Synthesis - to answer the questions, which the audience seeks in text-form very short and sharp.


## 5. Visualizations

### 5.1 Explain the chosen visualizations

In order to answer who the target group is following visualizations are used: 
1. Bookeh interactive pie chart 
2. Folium map
3. Bookeh interactive Pie chart combined with ipywidgets
4. Bookeh interactive vbar combined with ipywidgets

### 5.2 Why they are right for the story to tell

In order to find where to sell the product and to whom, the above-mentioned visualizations are used. They are all interactive in their own ways. 

The 1th visualization explains the majority of ownership of building in Copenhagen. Pie chart is used to easily visualize and compare whether companies or private owners owns the buildings in majority. 

The 2th visualization is a map, that shows the heating consumption per square meter. When working with cities and values related to cities maps gives a good overview. They also have the possibility to interact with the data by zooming, filtering and hovering. 

The 3th visualization is an interactive pie chart combined with ipywidgets and it's used to visualize the comparison of different civil conditions. Ipywidgets is used, so that the user can select the city and investigate closer. 

The 4th visualization is an interactive vbar combined with ipywidgets and it is used to visualize the different group of ages and their number of persons. Ipywidgets is used, so that the user can select the city and investigate closer.
